```python
import pandas as pd
```

```python
cp = pd.read_csv('/content/cancer preduction.csv.txt')
```

```python
cp.head(2)
```

```python
cp.info()
```

```python
cp.describe()
```

```python
cp.columns
```

```python
y=cp['diagnosis']
```

```python
X = cp.drop(['id','diagnosis','Unnamed: 32','concavity_se'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.8, random_state=2020)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=5000)
```

```python
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test, y_test)
```

```python
accuracy_score(y_test, y_pred)
```
